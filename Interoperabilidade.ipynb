```python
htlc = pd.read_csv('htlc.csv')
htlc.head()
```

```python
newsheet  = htlc.drop(columns=['','','',''])
newsheet.head()
```

```python
plt.scatter(htlc['date'],htlc['timeMintArbitrum']/1000,  label= 'Mint Arbitrum')
plt.scatter(htlc['date'],htlc['timeApproveArbitrum']/1000, label='Approve Arbitrum')
plt.scatter(htlc['date'],htlc['timeTransferArbitrum']/1000, label='Transfer Arbitrum')
plt.scatter(htlc['date'],htlc['timeMintAmoy']/1000, label='Mint Amoy')
plt.xlabel('')
plt.ylabel('Tempo(s)')
plt.legend(loc='upper right')
plt.title('tempo')
plt.show()
```
